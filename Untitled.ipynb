```python
import os
import re
from PyPDF2 import PdfReader

# ========= CONFIG =========
CV_PATH = "dataset/CV_Sanjib_Shah.pdf"   

# ========= LOAD CV =========

def parse_pdf(path: str) -> str:
    """Extract text from a PDF file."""
    reader = PdfReader(path)
    text = ""

    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text + "\n"

    return text

# Example usage



full_text = parse_pdf(CV_PATH)
print(full_text[:5200]) 
# ========= SECTION SPLITTER =========
SECTION_HEADERS = [
    "ABOUT ME",
    "WORK EXPERIENCE",
    "EDUCATION AND TRAINING",
    "PROJECTS",
    "VOLUNTEERING",
    "PUBLICATIONS",
    "CONFERENCES AND SEMINARS",
    "HONOURS AND AWARDS",
    "SKILLS"
]

def split_by_sections(text):
    sections = {}
    current_section = "BASIC INFO"
    buffer = []

    lines = text.splitlines()
    for line in lines:
        line_clean = line.strip()
        if line_clean.upper() in SECTION_HEADERS:
            sections[current_section] = "\n".join(buffer).strip()
            current_section = line_clean.upper()
            buffer = []
        else:
            buffer.append(line)
    sections[current_section] = "\n".join(buffer).strip()
    return sections

sections = split_by_sections(full_text)




print("finish")
```
